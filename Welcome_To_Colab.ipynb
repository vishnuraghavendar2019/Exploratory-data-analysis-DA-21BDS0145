```
import numpy as np
# 1.1 Classify Batsmen (Defensive, Radical, Brute) based on format-specific strike rates

def classify_batsman(format_type, strike_rate):
    """
    Categorizes batsman as Defensive, Radical, or Brute based on strike rate & format.
    """
    if format_type == 'T20':
        if strike_rate <= 125:
            return 'Defensive'
        elif strike_rate <= 150:  # implicitly > 125
            return 'Radical'
        else:
            return 'Brute'
    elif format_type == 'ODI':
        if strike_rate <= 100:
            return 'Defensive'
        elif strike_rate <= 125:  # implicitly > 100
            return 'Radical'
        else:
            return 'Brute'
    elif format_type == 'Test':
        if strike_rate < 50:
            return 'Defensive'
        elif strike_rate <= 100:  # implicitly >= 50
            return 'Radical'
        else:
            return 'Brute'
    return 'Unknown'  # fallback

# Apply the classification to each dataset
t20_bat['BatCategory'] = t20_bat['Strike rate'].apply(lambda x: classify_batsman('T20', x))
odi_bat['BatCategory'] = odi_bat['Strike rate'].apply(lambda x: classify_batsman('ODI', x))
test_bat['BatCategory'] = test_bat['Strike rate'].apply(lambda x: classify_batsman('Test', x))

# 1.2 Classify Bowlers or identify bowlers by style (pace, spin) if available
# For simplicity, assume we have a 'BowlingStyle' column or define it with your own logic
# e.g., we can do a placeholder if not in dataset:
if 'BowlingStyle' not in t20_bowl.columns:
    t20_bowl['BowlingStyle'] = 'Pace'  # Example fallback
if 'BowlingStyle' not in odi_bowl.columns:
    odi_bowl['BowlingStyle'] = 'Spin'  # Example fallback
if 'BowlingStyle' not in test_bowl.columns:
    test_bowl['BowlingStyle'] = 'Pace' # Example fallback

# 1.3 Decide team composition by format
# T20: 5-6 batsmen, 3-4 bowlers, 1-2 all-rounders
# ODI: 6-7 batsmen, 4 bowlers, 1-2 all-rounders
# Test: 6-7 batsmen, 4-5 bowlers, 1 all-rounder

# We'll define these as a dictionary for reference in Step 6:
TEAM_COMPOSITION = {
    'T20':  {'batsmen': 5, 'bowlers': 4, 'allrounders': 2},
    'ODI':  {'batsmen': 6, 'bowlers': 4, 'allrounders': 1},
    'Test': {'batsmen': 6, 'bowlers': 4, 'allrounders': 1}
}

print("Step 1 completed: Player roles and team composition defined.")
```

```
# ------------------
# STEP 2: ASSIGN WEIGHTS
# ------------------

# Example weighting:
# We give numeric weights to each batting category depending on format
WEIGHTS_BAT = {
    'T20': {
        'Defensive': 0.8,
        'Radical': 1.0,
        'Brute': 1.2
    },
    'ODI': {
        'Defensive': 1.0,
        'Radical': 1.1,
        'Brute': 1.1
    },
    'Test': {
        'Defensive': 1.2,
        'Radical': 1.0,
        'Brute': 0.9
    }
}

# For bowlers, consider economy or average important for T20, wicket-taking for Test, etc.
# We'll define a simple approach: pace vs. spin, etc.
# (Expand logic as needed)
WEIGHTS_BOWL = {
    'T20': {
        'Pace': 1.1,
        'Spin': 1.0
    },
    'ODI': {
        'Pace': 1.0,
        'Spin': 1.0
    },
    'Test': {
        'Pace': 1.0,
        'Spin': 1.1
    }
}

# We'll use these weight dictionaries in Step 3 (Impact Score) to multiply relevant stats

print("Step 2 completed: Weights assigned to batting categories and bowling styles.")
print("WEIGHTS_BAT:\n", WEIGHTS_BAT)
print("\nWEIGHTS_BOWL:\n", WEIGHTS_BOWL)
```

```
# STEP 3: Calculate Player Impact Scores
###############################################################################
def batting_impact_score(row, format_type):
    """
    (Strike Rate * Average * Runs) / 1000, multiplied by category weight
    """
    sr = row.get('Strike rate', 0.0)
    avg = row.get('Average', 0.0)
    runs = row.get('Runs', 0.0)
    cat = row.get('BatCategory','Defensive')
    base_score = (sr * avg * runs)/1000 if runs>0 else 0
    weight = WEIGHTS_BAT[format_type].get(cat,1.0)
    return base_score * weight

odi_bowl['Economy'] = pd.to_numeric(odi_bowl['Economy'], errors='coerce')
odi_bowl['Average'] = pd.to_numeric(odi_bowl['Average'], errors='coerce')
odi_bowl['Wkts'] = pd.to_numeric(odi_bowl['Wkts'], errors='coerce')

test_bowl['Economy'] = pd.to_numeric(test_bowl['Economy'], errors='coerce')
test_bowl['Average'] = pd.to_numeric(test_bowl['Average'], errors='coerce')
test_bowl['Wkts'] = pd.to_numeric(test_bowl['Wkts'], errors='coerce')

odi_bowl['BowlImpact'] = odi_bowl.apply(lambda row: bowling_impact_score(row, 'ODI'), axis=1)

def bowling_impact_score(row, format_type):
    """
    (Wickets * (1/Economy) * (1/Average)) * 100, multiplied by style weight
    """
    wkts = row.get('Wkts',0.0)
    eco = row.get('Economy',7.0)
    bowl_avg = row.get('Average',25.0)
    style = row.get('BowlingStyle','Pace')

    base_score = 0
    if eco>0 and bowl_avg>0:
        base_score = wkts*(1/eco)*(1/bowl_avg)*100

    weight = WEIGHTS_BOWL[format_type].get(style, 1.0)
    return base_score*weight

# Apply to T20
t20_bat['BatImpact'] = t20_bat.apply(lambda row: batting_impact_score(row, 'T20'), axis=1)
t20_bowl['BowlImpact'] = t20_bowl.apply(lambda row: bowling_impact_score(row, 'T20'), axis=1)

# Similarly for ODI, Test
odi_bat['BatImpact'] = odi_bat.apply(lambda row: batting_impact_score(row, 'ODI'), axis=1)
odi_bowl['BowlImpact'] = odi_bowl.apply(lambda row: bowling_impact_score(row, 'ODI'), axis=1)
test_bat['BatImpact'] = test_bat.apply(lambda row: batting_impact_score(row, 'Test'), axis=1)
test_bowl['BowlImpact'] = test_bowl.apply(lambda row: bowling_impact_score(row, 'Test'), axis=1)
print("STEP 3 COMPLETE: Player Impact Scores calculated.")
print("\nT20 Bat after classification and BatImpact:\n", t20_bat[['Player','BatCategory','BatImpact']])
print("\nT20 Bowl after BowlImpact:\n", t20_bowl[['Player','BowlingStyle','BowlImpact']])
```

```
# STEP 4: Filter Players Based on Match Conditions
###############################################################################
def filter_players_by_conditions(bat_df, bowl_df, pitch_type, part_of_day, dew):
    # existing code logic...
    pass
    """
    Example condition-based approach. Adjust as you like.
    """
    filtered_bat = bat_df.copy()
    filtered_bowl = bowl_df.copy()

    # Pitch-based logic
    pitch_lower = pitch_type.lower()
    if 'bat' in pitch_lower or 'flat' in pitch_lower:
        # bonus for brute/radical in batting
        filtered_bat['ConditionBonus'] = np.where(
            filtered_bat['BatCategory'].isin(['Brute','Radical']), 1.1, 1.0
        )
    else:
        filtered_bat['ConditionBonus'] = 1.0

    # Dew logic => bowlers have an additional challenge
    if dew.lower() == 'yes':
        filtered_bowl['DewBonus'] = 1.1
    else:
        filtered_bowl['DewBonus'] = 1.0

    # Part of day => e.g. morning helps seam
    if part_of_day.lower() == 'morning':
        filtered_bowl['TimeBonus'] = np.where(filtered_bowl['BowlingStyle']=='Pace', 1.2, 1.0)
    else:
        filtered_bowl['TimeBonus'] = 1.0

    return filtered_bat, filtered_bowl

# Try filtering with an example
pitch_type = 'batting-friendly'
part_of_day = 'Evening'
dew = 'Yes'
filtered_bat, filtered_bowl = filter_players_by_conditions(
    t20_bat, t20_bowl, pitch_type, part_of_day, dew
)

print("STEP 4 COMPLETE: Filtered based on pitch='batting-friendly', part_of_day='Evening', dew='Yes'.")
print("\nFiltered Batsmen sample:\n", filtered_bat[['Player','BatCategory','ConditionBonus']])
print("\nFiltered Bowlers sample:\n", filtered_bowl[['Player','BowlingStyle','DewBonus','TimeBonus']])
```

```
# STEP 5: Rank Players
###############################################################################
def rank_players(bat_df, bowl_df):
    """
    Combine impact scores & condition bonuses, scale 0-100 with MinMaxScaler.
    """
    if 'ConditionBonus' not in bat_df.columns:
        bat_df['ConditionBonus'] = 1.0
    bat_df['FinalBatScore'] = bat_df['BatImpact']*bat_df['ConditionBonus']

    scaler_bat = MinMaxScaler((0,100))
    bat_df[['ScaledBatScore']] = scaler_bat.fit_transform(bat_df[['FinalBatScore']])
    bat_df.sort_values('ScaledBatScore', ascending=False, inplace=True)
    bat_df.reset_index(drop=True, inplace=True)

    if 'DewBonus' not in bowl_df.columns:
        bowl_df['DewBonus'] = 1.0
    if 'TimeBonus' not in bowl_df.columns:
        bowl_df['TimeBonus'] = 1.0
    bowl_df['FinalBowlScore'] = bowl_df['BowlImpact']*bowl_df['DewBonus']*bowl_df['TimeBonus']

    scaler_bowl = MinMaxScaler((0,100))
    bowl_df[['ScaledBowlScore']] = scaler_bowl.fit_transform(bowl_df[['FinalBowlScore']])
    bowl_df.sort_values('ScaledBowlScore', ascending=False, inplace=True)
    bowl_df.reset_index(drop=True, inplace=True)

    return bat_df, bowl_df

ranked_bat, ranked_bowl = rank_players(filtered_bat, filtered_bowl)

print("STEP 5 COMPLETE: Players ranked and scaled to 0-100.")
print("\nRanked Batsmen:\n", ranked_bat[['Player','FinalBatScore','ScaledBatScore']])
print("\nRanked Bowlers:\n", ranked_bowl[['Player','FinalBowlScore','ScaledBowlScore']])
```

```
# STEP 6: Select Playing XI
###############################################################################
def select_playing_xi(bat_df, bowl_df, format_type):
    comp = TEAM_COMPOSITION[format_type]
    n_bat = comp['batsmen']
    n_bowl = comp['bowlers']

    selected_batsmen = bat_df.head(n_bat)
    selected_bowlers = bowl_df.head(n_bowl)

    # Combine; if under 11, pick next best from whichever is highest
    final_team = pd.concat([selected_batsmen, selected_bowlers]).drop_duplicates('Player')
    while len(final_team)<11:
        next_bat = bat_df[~bat_df['Player'].isin(final_team['Player'])].head(1)
        next_bowl = bowl_df[~bowl_df['Player'].isin(final_team['Player'])].head(1)
        if not next_bat.empty and not next_bowl.empty:
            if next_bat['ScaledBatScore'].iloc[0] > next_bowl['ScaledBowlScore'].iloc[0]:
                final_team = pd.concat([final_team, next_bat])
            else:
                final_team = pd.concat([final_team, next_bowl])
        elif not next_bat.empty:
            final_team = pd.concat([final_team, next_bat])
        elif not next_bowl.empty:
            final_team = pd.concat([final_team, next_bowl])
        else:
            break
    final_team = final_team.head(11).reset_index(drop=True)
    return final_team

final_xi = select_playing_xi(ranked_bat, ranked_bowl, format_type='T20')

print("STEP 6 COMPLETE: Playing XI selected.")
print("\nFinal XI:\n", final_xi[['Player','BatCategory','BowlingStyle','ScaledBatScore','ScaledBowlScore']])
```

```
###############################################################################
# STEP 7: Provide Reasoning
###############################################################################
def provide_reasoning(final_team):
    reasons = []
    for _, row in final_team.iterrows():
        if 'BatCategory' in row and pd.notna(row['BatCategory']):
            reasons.append(f"{row['Player']}: {row['BatCategory']} batsman, scaled score = {row.get('ScaledBatScore',0):.2f}")
        elif 'BowlingStyle' in row and pd.notna(row['BowlingStyle']):
            reasons.append(f"{row['Player']}: {row['BowlingStyle']} bowler, scaled score = {row.get('ScaledBowlScore',0):.2f}")
        else:
            reasons.append(f"{row['Player']}: Key player")
    return reasons

reasons_list = provide_reasoning(final_xi)
print("STEP 7 COMPLETE: Reasoning for each selected player.")
for reason in reasons_list:
    print(reason)
```

```
# STEP 8: Example Stadium Data & Toss Decision
stadium_data = pd.DataFrame({
    'StadiumName': ['Wankhede Stadium','Lord\'s','MCG','Wanderers','Eden Park'],
    'Country': ['India','England','Australia','South Africa','New Zealand'],
    'SupportInnings1': ['Bowl','Bowl','Bat','Bowl','Bowl'],
    'SupportInnings2': ['Bat','Bat','Bat','Bat','Bat']
})

def decide_toss(stadium_name, dew, pitch):
    # If dew => prefer Bowl
    if dew.lower() == 'yes':
        return 'Bowl'
    # Otherwise, lookup stadium
    row = stadium_data[stadium_data['StadiumName']==stadium_name]
    if row.empty:
        return 'Bat'  # fallback
    # If pitch is batting-friendly, maybe prefer bat
    if 'bat' in pitch.lower():
        return 'Bat'
    else:
        return row.iloc[0]['SupportInnings1']  # simplistic approach

# Example usage
stadium_name = 'Wankhede Stadium'
dew = 'Yes'
pitch = 'batting-friendly'

toss_decision = decide_toss(stadium_name, dew, pitch)
print("STEP 8 COMPLETE: Stadium & Toss Example")
print(f"Stadium: {stadium_name}, Dew: {dew}, Pitch: {pitch}")
print("Toss Decision =>", toss_decision)
```
